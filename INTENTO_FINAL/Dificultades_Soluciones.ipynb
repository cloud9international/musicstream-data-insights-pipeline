### **DIFICULTADES, SOLUCIONES Y POSIBLES IMPLEMENTACIONES**

**LINKS DE LA BIO**
- PROBLEMA: la biografia nos salía con un enlace y nosotras queríamos el texto
- SOLUCIÓN: crear un código que entarar a ese enlace y nos devolviera el texto. Después, crear una función para limpiar las celdas que seguian devolviendo un link y aplicarlo en un bucle for.

**VALORES NULOS**
- PROBLEMA: datos con valores Nulos
- SOLUCIÓN: convertir Nulos de NaN a None, para que SQL acepte los datos

**DUPLICADOS SPOTIFY**
- PROBLEMA: cargar los datos de Spotify en SQ - 61 IDs repetidos.
- SOLUCIÓN: código que al insertar los datos, si encuentra un ID repetido no lo incluye y nos devuelve un mensaje de error con el ID repetido. Pero sigue cargando los demás datos.
