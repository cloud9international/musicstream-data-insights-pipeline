Problema para cargar los datos de Spotify en SQL: DUPLICADOS - 61 IDs repetidos
Solución: código que al insertar los datos si encuentra un ID repetido no lo incluye y nos devuelve un mensaje de error con el ID repetido. Pero sigue cargando los demás datos.
