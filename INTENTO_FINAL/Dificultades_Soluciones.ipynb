**Preguntas que necesitamos subconsultas**

- ¿Cuál es el artista que tiene una popularidad promedio en Spotify superior al promedio de popularidad general de todas las canciones en la base de datos?

```python
SELECT artist_name, AVG(popularity) AS popularidad_promedio
    FROM spotify
    GROUP BY 1
    HAVING AVG(popularity) > (SELECT AVG(popularity)
                                FROM spotify);
```

- Identifica las canciones más populares (máxima popularity por género) y muestra el nombre de la canción, el artista y el género.

```python
SELECT
  T1.track_name,
  T1.artist_name,
  T1.genre,
  T1.popularity
    FROM spotify AS T1
    INNER JOIN (SELECT genre, MAX(popularity) AS max_popularity
                  FROM spotify
                GROUP BY genre) AS T2
        ON T1.genre = T2.genre AND T1.popularity = T2.max_popularity;
```
