```python
Soundtracks = {}

for year in range(2015, 2020):   # 2020 no incluido
    print(f"\nüéµ 50 AFRO SONGS FROM {year}")
    print("_____________________________________")

    datos = sp.search(q=f"genre:afro year:{year}", limit=50, offset=0)  # B√∫squeda en Spotify


    tracks = datos["tracks"]["items"]

    for track in tracks:
        print(track['name'], "-",
            "Artist: ", track['artists'][0]['name'],
            "Track ID: ", track['id'],
            "Album type: ", track['album']['album_type'],
            "Release date: ", track['album']['release_date'],
            "Popularity", track['popularity'])

        Soundtracks[track['name']] = {
            'track': track['id'],
            'artist': track['artists'][0]['name'],
            'album_type': track['album']['album_type'],
            'release_date': track['album']['release_date'],
            'popularity': track['popularity']
        }
```
