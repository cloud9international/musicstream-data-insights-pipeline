```python
import pandas as pd
import requests, time, re, html

# ‚öôÔ∏è Reemplaza con tu clave de la API de Last.fm
API_KEY = "acb31d7ca15f87cc27d929d29fc8c8d8 "
LANG = "en"   # o "en" si prefieres biograf√≠as en ingl√©s

def quitar_html(x: str) -> str:
    """Quita etiquetas HTML y caracteres especiales."""
    if not x:
        return None
    x = re.sub(r"<.*?>", "", x)
    return html.unescape(x).strip() or None

def get_bio_lastfm(artist):
    """Pide la biograf√≠a del artista desde la API."""
    url = "https://ws.audioscrobbler.com/2.0/"
    params = {
        "method": "artist.getInfo",
        "artist": artist,
        "api_key": API_KEY,
        "format": "json",
        "autocorrect": 1,
        "lang": LANG
    }
    r = requests.get(url, params=params, timeout=15)
    if r.status_code != 200:
        print(f"‚ùå Error HTTP {r.status_code} para {artist}")
        return None, None
    data = r.json()
    bio = data.get("artist", {}).get("bio", {})
    summary = quitar_html(bio.get("summary"))
    content = quitar_html(bio.get("content"))
    return summary, content

# üß† 1Ô∏è‚É£ Carga el CSV de prueba
df = pd.read_csv("artistas_lastfm_test_36_test_limpiar_bios.csv")

# üß† 2Ô∏è‚É£ Crea nuevas columnas vac√≠as
df["bio_summary_txt"] = None
df["bio_content_txt"] = None

# üß† 3Ô∏è‚É£ Itera y obtiene biograf√≠as
for i, artist in enumerate(df["artist_name"], start=1):
    summary, content = get_bio_lastfm(artist)
    df.loc[i-1, "bio_summary_txt"] = summary
    df.loc[i-1, "bio_content_txt"] = content
    print(f"{i}. ‚úÖ {artist}")
    time.sleep(0.25)   # pausa entre peticiones

# üß† 4Ô∏è‚É£ Muestra primeros resultados
df.head()
```
