```python
def get_artist_info(artist_name):
    API_KEY = my_API_KEY 
    BASE_URL = "https://ws.audioscrobbler.com/2.0/"

    params = {
        "method": "artist.getinfo",
        "artist": artist_name,
        "api_key": API_KEY,
        "format": "json",
        "autocorrect": 1,
        "lang": "es"
    }

    response = requests.get(BASE_URL, params=params)
    data = response.json()

    # Validar si hay error
    if "error" in data:
        print(f"‚ùå Error: {data['message']}")
        return None

    artist = data["artist"]
    bio = artist["bio"]["summary"]
    listeners = artist["stats"]["listeners"]
    playcount = artist["stats"]["playcount"]
    similar_artists = [a["name"] for a in artist["similar"]["artist"]]

    return {
        "nombre": artist["name"],
        "biografia": bio,
        "oyentes": listeners,
        "reproducciones": playcount,
        "similares": similar_artists
    }


# ============================
# üß† Ejemplo de uso
# ============================

if __name__ == "__main__":
    artista = input("üéµ Ingresa el nombre del artista: ")
    info = get_artist_info(artista)

    if info:
        print("\nüé§ Artista:", info["nombre"])
        print("üìñ Biograf√≠a:\n", info["biografia"])
        print("üë• Oyentes:", info["oyentes"])
        print("‚ñ∂Ô∏è Reproducciones:", info["reproducciones"])
        print("üé∂ Artistas similares:", ", ".join(info["similares"]))
```
